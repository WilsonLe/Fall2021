```python
# *** Remove life and cell columns from table ***
incGDP = indicatorsDF.copy()
incGDP.drop(["life", "cell"], axis=1, inplace=True)
incGDP.head(10)
```

```python
# *** Remove rows that do not have valid population or gdp entries ***

incGDPtemp = incGDP.copy()   # create tempory copy of table
drops = []                   # indices of rows that should be removed 
for rowlabel, rowseries in incGDP.iterrows():
    if pd.isna(rowseries["gdp"]) or pd.isna(rowseries["pop"]):
        drops.append(rowlabel)                                  # add index is entry(s) is NaN
        
incGDPtemp.drop(drops, axis = 0, inplace=True)  # remove entries from drops
incGDP = incGDPtemp
incGDP.head(10)
```

```python
# ** Add GDP per column ***
incGDP2 = incGDP.copy()
gdpper = incGDP["gdp"]/incGDP["pop"]
incGDP2["gdp-per"]= gdpper
incGDP2.head(10)
```

```python
# *** Build dictionary and plot histogram ***
import matplotlib.pyplot as pyplot

gdpPerPopDict = {}  # dictionary that will map counties to average gdp
curCountry = "Aruba"
total = 0 
count = 0
for rowlabel, rowseries in incGDP2.iterrows():           # loop through rows in table 
    if curCountry != rowseries["country"]:               # if current country is new...
        if total/count  > 20:                            # reset values, add current value to dict 
            gdpPerPopDict[curCountry] = total/count
        total = 0
        count = 0
        curCountry = rowseries["country"]
    total += rowseries["gdp-per"]                       # update values for current country
    count += 1
    
pyplot.bar(gdpPerPopDict.keys(), gdpPerPopDict.values())
pyplot.xticks(range(len(gdpPerPopDict.keys())), rotation = 270, fontsize = 'small')
pyplot.show()
```
