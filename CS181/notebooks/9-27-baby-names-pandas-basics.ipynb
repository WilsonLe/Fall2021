```python
# Column indices used in topnames.csv
_year = 0
_gender = 1
_name = 2
_count = 3
_filename = 'topnames.csv'  

def buildDol():
    '''Build a dictionary of lists representation of table from the file'''
    file = open("topnames.csv", 'r')
    dol = {}
    columns = file.readline().strip().split(',')    # read first line
    
    for field in columns:   # initialize mapping of column fields
        dol[field] = []
        
    for line in file:
        curRow = line.strip().split(',')
        dol[columns[_year]].append(int(curRow[_year]))
        dol[columns[_gender]].append(curRow[_gender])
        dol[columns[_name]].append(curRow[_name])
        dol[columns[_count]].append(int(curRow[_count]))
  
    return dol
```

```python
# create a "data frame" object in python from pandas module
import pandas as pd 

babyNamesDol = buildDol()
babyDF = pd.DataFrame(babyNamesDol)

babyDF # display table in markdown 
```
