```python
import pandas as pd
fileName = "indicators.csv"
indicatorsDF = pd.read_csv(fileName) # read file directly from .csv
indicatorsDF.head(10)
```

```python
# *** Deleting a returning a single column ***

indicatorsCopy =  indicatorsDF.copy()
indicatorsCopy.set_index(["code", "year"], inplace=True)
gdpSeries = indicatorsCopy.pop("gdp") # use pop() method, and pass in column name
indicatorsCopy.head(10)
```

```python
gdpSeries # pop() returns series object containing the object
```

```python
#  ** Deleting multiple columns  ***
indicatorsCopy =  indicatorsDF.copy() 
indicatorsCopy.set_index(["code", "year"], inplace=True)

# deletes column labels list passed in. 
# optional parameter axis=1 corresponds to deleting columns
indicatorsCopy.drop(["cell", "life"], axis =1, inplace=True) 
indicatorsCopy
```

```python
#  *** Deleting multiple columns  ***

indicatorsCopy =  indicatorsDF.copy()
indicatorsCopy.set_index(["code"], inplace=True)

# deletes row labels list passed in. 
# optional parameter axis=0 corresponds to deleting rows
indicatorsCopy.drop(["ZWE"], axis=0, inplace=True) 
indicatorsCopy.tail()
```

```python
# *** Adding a column to the table *** 

indicatorsCopy =  indicatorsDF.copy()

gdpGrowth = indicatorsCopy["gdp"] * 1.15  # column if gdp grows by 15%
indicatorsCopy["gdp growth"] = gdpGrowth
indicatorsCopy
```

```python
# *** Alternate method for adding column to the table *** 

indicatorsCopy =  indicatorsDF.copy()
for rowlabel, rowseries in indicatorsDF.iterrows():
    indicatorsCopy.loc[rowlabel, "gdp"] = indicatorsCopy.loc[rowlabel, "gdp"]*1.15
    
indicatorsCopy
```
