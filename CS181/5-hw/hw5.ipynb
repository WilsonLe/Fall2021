### Data Overview

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

worldBankDf = pd.read_csv("world_bank_pop.csv")
worldBankDf.head(10)
```

The above table includes population data for world countries for the years 2000 through 2017 (from S&P World Bank database). The entries in each row correspond to a country, a population statistic, and then the remaining entries list the values of that statistic for each year from 2000 to 2017. For each country, there are four statistics listed (again, corresponding to their own row):

* SP.POP.TOTL : Total population of the country.
* SP.POP.GROW : Population growth from the previous year (percentage growth).
* SP.URB.TOTL : Total population for urban areas of the country.
* SP.URB.GROW : Population growth of urban areas (percentage growth).

## Part 1 [10 points]

Using table transformation operators we saw in class, to create a data frame that is tidy with respect to the following mapping:

$$ \text{year, country} \rightarrow \text{SP.POP.TOTL, SP.POP.GROW, SP.URB.TOTL , SP.URB.GROW}$$

Note that it's possible to do pivots on a list of indices (instead of a single index as we saw in class), but you should use the <code>pivot_table()</code> function instead - see page 268 in the textbook for an example. Also, be sure that your tansformed data frame is indexed by (year, country). 

```python
import numpy as np

worldBankDfMelt = worldBankDf.melt(id_vars=["country", "indicator"], var_name="year", value_name="population")
worldBankDfMelt.head(10)

worldBankDfPivot = worldBankDfMelt.pivot_table(index=["year", "country"], columns="indicator", values="population")
worldBankDfPivot
# worldBankDfPivot.index
```

## Part 2 [15 points] 

Let's define the *rural population* of a country to be the total population that is not included in the urban subtotal. Given this defintion, update your transformed data frame so that it now includes two new columns with headers: SP.RURL.TOTL and SP.RURL.GROW, that (respectively) corresponds to the total rural population and the rural population growth from the previous year. A couple things to note:

* For any (year, country) pair in the table that has NaNs for either total or urban population, you should put in NaNs for your newly created SP.RURL.TOTL and SP.RURL.GROW entries. To create an NaN value in python, use the function call <code>float("nan")</code>. 
* Given that there is no explict data for 1999, you will have to bit a more clever when determining the rural growth for the 2000 entries.

```python
worldBankDfPivot['SP.RURL.TOTL'] = worldBankDfPivot['SP.POP.TOTL'] - worldBankDfPivot['SP.URB.TOTL']
worldBankDfPivot['SP.RURL.GROW'] = worldBankDfPivot[ 'SP.POP.GROW'] - worldBankDfPivot['SP.URB.GROW']
worldBankDfPivot.dropna(inplace=True)
worldBankDfPivot
```

## Part 3 [15 points] 

Using the data frame you created from Part 2, construct a .json file that converts the above table into a hierarchical representation. In particular, the tree represented by your .json file should have four levels of branching. At the top level, the branching from the root is based on years (i.e., one subtree for 2000, a subtree for 2001, etc.). At the next subtree down, the tree then branches based on country. Then for each country subtree, there are branches each of the total, urban, and rural population data categoreis. Then at these bottom level subtrees, are two leaves that store both the total and growth in each of these subtrees category. Again, the result of your code should be a .json file that is written to your working directory.  

```python
from pprint import pprint
dod = {}

for year, country in worldBankDfPivot.index:
	if year not in dod.keys():
		dod[year] = {}
	if country not in dod.keys():
		dod[year][country] = {}
	for key in worldBankDfPivot.loc[(year, country)].index:
		if key == "SP.POP.GROW":
			if "POP" not in dod[year][country].keys():
				dod[year][country]["POP"] = {}
			dod[year][country]["POP"]["GROW"] = worldBankDfPivot.loc[(year, country)][key]
		elif key == "SP.POP.TOTL":
			if "POP" not in dod[year][country].keys():
				dod[year][country]["POP"] = {}
			dod[year][country]["POP"]["TOTL"] = worldBankDfPivot.loc[(year, country)][key]
		elif key == "SP.URB.GROW":
			if "URB" not in dod[year][country].keys():
				dod[year][country]["URB"] = {}
			dod[year][country]["URB"]["GROW"] = worldBankDfPivot.loc[(year, country)][key]
		elif key == "SP.URB.TOTL":
			if "URB" not in dod[year][country].keys():
				dod[year][country]["URB"] = {}
			dod[year][country]["URB"]["TOTL"] = worldBankDfPivot.loc[(year, country)][key]
		elif key == "SP.RURL.GROW":
			if "RURL" not in dod[year][country].keys():
				dod[year][country]["RURL"] = {}
			dod[year][country]["RURL"]["GROW"] = worldBankDfPivot.loc[(year, country)][key]
		elif key == "SP.RURL.TOTL":
			if "RURL" not in dod[year][country].keys():
				dod[year][country]["RURL"] = {}
			dod[year][country]["RURL"]["TOTL"] = worldBankDfPivot.loc[(year, country)][key]
```

I calculate the average population, rural, and urban population/growth of all the countries from 2000 to 2019. I then extract those into a list and sort then based on the total population. I then extract the lowest 10 population countries and perform graphing and analysis on them.

The 10 countries with the lowest average population were ['TUV', 'NRU', 'PLW', 'VGB', 'TCA', 'SMR', 'GIB', 'SXM', 'LIE', 'MCO']. There does not seems to be a trend between average population and average population growth. The countries with the lowest average population is different to the countries with the lowest average growth. Between those countries rural and urban population and growth, it seems like the majority of them are urban population, with the exception of LIE.



```python
# calculate average total of each country over the year
numYear = len(dod)

# transform original
dodCountryYear = {}

for countryCode in dod["2000"].keys():
	dodCountryYear[countryCode] = {}
for year, country in dod.items():
	for countryCode, populationData in country.items():
		dodCountryYear[countryCode][year] = {}
		dodCountryYear[countryCode][year]["POP"] = {
			"GROW": populationData["POP"]["GROW"],
			"TOTL": populationData["POP"]["TOTL"],
		}
		dodCountryYear[countryCode][year]["URB"] = {
			"GROW": populationData["URB"]["GROW"],
			"TOTL": populationData["URB"]["TOTL"],
		}
		dodCountryYear[countryCode][year]["RURL"] = {
			"GROW": populationData["RURL"]["GROW"],
			"TOTL": populationData["RURL"]["TOTL"],
		}


countryPopTotlAverage = [] # lod containing country population averages
countryPopGrowAverage = [] # lod containing country population grow averages
countryRurlTotlAverage = [] # lod containing country rural averages
countryRurlGrowAverage = [] # lod containing country rural grow averages
countryUrbTotlAverage = [] # lod containing country urban averages
countryUrbGrowAverage = [] # lod containing country urban grow averages

for country, yearData in dodCountryYear.items():
	popTotl = 0
	popGrow = 0
	rurlTotl = 0
	rurlGrow = 0
	urbTotl = 0
	urbGrow = 0
	for year, data in yearData.items():
		popTotl += data["POP"]["TOTL"]
		popGrow += data["POP"]["GROW"]
		rurlTotl += data["RURL"]["TOTL"]
		rurlGrow += data["RURL"]["GROW"]
		urbTotl += data["URB"]["TOTL"]
		urbGrow += data["URB"]["GROW"]

	countryPopTotlAverage.append({
		"country": country,
		"popTotl": popTotl/numYear,
	})
	countryPopGrowAverage.append({
		"country": country,
		"popGrow": popGrow/numYear
	})
	countryRurlTotlAverage.append({
		"country": country,
		"rurlTotl": rurlTotl/numYear
	})
	countryRurlGrowAverage.append({
		"country": country,
		"rurlGrow": rurlGrow/numYear
	})
	countryUrbTotlAverage.append({
		"country": country,
		"urbTotl": urbTotl/numYear
	})
	countryUrbGrowAverage.append({
		"country": country,
		"urbGrow": urbGrow/numYear
	})

countryPopTotlAverage.sort(key = lambda x: x["popTotl"])
countryPopGrowAverage.sort(key = lambda x: x["popGrow"])
countryRurlTotlAverage.sort(key = lambda x: x["rurlTotl"])
countryRurlGrowAverage.sort(key = lambda x: x["rurlGrow"])
countryUrbTotlAverage.sort(key = lambda x: x["urbTotl"])
countryUrbGrowAverage.sort(key = lambda x: x["urbGrow"])

lowestAvgPopCountries = countryPopTotlAverage[:10]
lowestAvgPopData = []
for country in lowestAvgPopCountries:
	code = country["country"]
	lowestAvgPopData.append({
		"country": code,
		"popTotl": country["popTotl"],
		"popGrow": next((x["popGrow"] for x in countryPopGrowAverage if x["country"] == code), None),
		"rurlTotl": next((x["rurlTotl"] for x in countryRurlTotlAverage if x["country"] == code), None),
		"rurlGrow": next((x["rurlGrow"] for x in countryRurlGrowAverage if x["country"] == code), None),
		"urbTotl": next((x["urbTotl"] for x in countryUrbTotlAverage if x["country"] == code), None),
		"urbGrow": next((x["urbGrow"] for x in countryUrbGrowAverage if x["country"] == code), None),
	})

def plot_total_population():
	labels = [x["country"] for x in lowestAvgPopData]
	x = np.arange(len(labels))  # the label locations
	popTotals = [x["popTotl"] for x in lowestAvgPopData]
	width = 0.35
	fig, ax = plt.subplots()
	rects1 = ax.bar(x - width/2, popTotals, width, label="Total Population")
	ax.set_xticks(x)
	ax.set_xticklabels(labels)
	ax.set_ylabel('Population')
	plt.title("total populations")
	plt.show()

def plot_population_grow():
	labels = [x["country"] for x in lowestAvgPopData]
	x = np.arange(len(labels))  # the label locations
	popGrow = [x["popGrow"] for x in lowestAvgPopData]
	width = 0.35
	fig, ax = plt.subplots()
	rects1 = ax.bar(x - width/2, popGrow, width, label="Total Population")
	ax.set_xticks(x)
	ax.set_xticklabels(labels)
	ax.set_ylabel('Growth')
	plt.title("population growth")
	plt.show()

def plot_urban_total():
	labels = [x["country"] for x in lowestAvgPopData]
	x = np.arange(len(labels))  # the label locations
	urbTotl = [x["urbTotl"] for x in lowestAvgPopData]
	width = 0.35
	fig, ax = plt.subplots()
	rects1 = ax.bar(x - width/2, urbTotl, width, label="Total Population")
	ax.set_xticks(x)
	ax.set_xticklabels(labels)
	ax.set_ylabel('Population')
	plt.title("Urban total")
	plt.show()

def plot_urban_grow():
	labels = [x["country"] for x in lowestAvgPopData]
	x = np.arange(len(labels))  # the label locations
	urbGrow = [x["urbGrow"] for x in lowestAvgPopData]
	width = 0.35
	fig, ax = plt.subplots()
	rects1 = ax.bar(x - width/2, urbGrow, width, label="Total Population")
	ax.set_xticks(x)
	ax.set_xticklabels(labels)
	ax.set_ylabel('Growth')
	plt.title("Urban growth")
	plt.show()

def plot_rural_total():
	labels = [x["country"] for x in lowestAvgPopData]
	x = np.arange(len(labels))  # the label locations
	rurlTotl = [x["rurlTotl"] for x in lowestAvgPopData]
	width = 0.35
	fig, ax = plt.subplots()
	rects1 = ax.bar(x - width/2, rurlTotl, width, label="Total Population")
	ax.set_xticks(x)
	ax.set_xticklabels(labels)
	ax.set_ylabel('Growth')
	plt.title("Rural total")
	plt.show()

def plot_rural_grow():
	labels = [x["country"] for x in lowestAvgPopData]
	x = np.arange(len(labels))  # the label locations
	rurlGrow = [x["rurlGrow"] for x in lowestAvgPopData]
	width = 0.35
	fig, ax = plt.subplots()
	rects1 = ax.bar(x - width/2, rurlGrow, width, label="Total Population")
	ax.set_xticks(x)
	ax.set_xticklabels(labels)
	ax.set_ylabel('Growth')
	plt.title("Rural growth")
	plt.show()

plot_total_population()
plot_population_grow()
plot_urban_total()
plot_urban_grow()
plot_rural_total()
plot_rural_grow()
```
