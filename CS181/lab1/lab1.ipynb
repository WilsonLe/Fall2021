## Part 1 [10 points] 
Write code in the cell below that reads in the <code>topnames.csv</code> file and creates a dictionary of lists (DoL) represenation of the table. Do not consult your textbook when writing your code (you may consult the example code from class).

```python
from pprint import pprint

def buildDoL(f):
    """build a dictionary of list from an input file

    Args:
            f (file): input csv file that has been read

    Returns:
            fields: list of fields in the input file
            DoL: dictionary of list that stores in put file, with keys are the fields in input file.
    """
    DoL = {}
    fields = f.readline().strip().split(',')
    for field in fields:
        DoL[field] = []
    for line in f:
        for i in range(len(fields)):
            value = line.strip().split(',')[i]
            DoL[fields[i]].append(value)
    return fields, DoL

filename = "topnames.csv"
f = open(filename, 'r')
fields, DoL = buildDoL(f)
print(fields)
pprint(DoL)
```
